```python
common_essentials = pd.read_csv("D:/Documents/Jayden Chronos test/RKI1_positivecontrol_genes.csv")["gene"]
nonessentials = pd.read_csv("D:/Documents/Jayden Chronos test/RKI1_negativecontrol_genes.csv")["gene"]

positive_controls = guide_map.sgrna[guide_map.gene.isin(common_essentials)]
negative_controls = guide_map.sgrna[guide_map.gene.isin(nonessentials)]

negative_controls.head(10)
```

```python
sequence_map['sequence_ID'] = sequence_map['sequence_ID'].astype(str)
sequence_map['cell_line_name'] = sequence_map['cell_line_name'].astype(str)
sequence_map['pDNA_batch'] = sequence_map['pDNA_batch'].astype(str)

guide_map['sgrna'] = guide_map['sgrna'].astype(str)
guide_map['gene'] = guide_map['gene'].astype(str)

readcounts = readcounts.astype(float)

positive_controls = positive_controls.astype(str)
negative_controls = negative_controls.astype(str)
```
